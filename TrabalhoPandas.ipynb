```python
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/DataFrames/votacao_secao_2022_DF.csv', encoding="ISO-8859-1", sep = ';')
df2 = pd.read_csv('/content/drive/My Drive/DataFrames/votacao_secao_2022_DF.csv', encoding="ISO-8859-1", sep = ';')
zonasEleitorais = pd.read_csv('/content/drive/My Drive/DataFrames/lista_zonas_eleitorais.csv', encoding="ISO-8859-1", sep = ',')

numeroZona = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11, 13, 14, 15 ,16, 17,18,19, 20, 21]

df2 = df2.drop_duplicates(subset='NM_VOTAVEL', keep='first')

print(" \n ===================== RESULTADO GERAL GOVERNADOR ====================== \n")
print("                              TOTAL de Votos: ", df.query("DS_CARGO == 'GOVERNADOR'")["QT_VOTOS"].sum())
data = [];
for item in df2.query("DS_CARGO == 'GOVERNADOR'")["NM_VOTAVEL"]:
  data.append([item, df.query("NM_VOTAVEL == '" + item + "' ")["QT_VOTOS"].sum()])
  resultado = pd.DataFrame(data, columns=['Nome Candidato', 'Quantidade de Votos'])


print("\n",resultado.sort_values(by='Quantidade de Votos', ascending=False));


print(" \n ====================== RESULTADO GERAL SENADOR ======================= \n")
print("                              TOTAL de Votos: ", df.query("DS_CARGO == 'SENADOR'")["QT_VOTOS"].sum())

data = [];
for item in df2.query("DS_CARGO == 'SENADOR'")["NM_VOTAVEL"]:
  data.append([item, df.query("NM_VOTAVEL == '" + item + "' ")["QT_VOTOS"].sum()])
  resultado = pd.DataFrame(data, columns=['Nome Candidato', 'Quantidade de Votos'])


print("\n",resultado.sort_values(by='Quantidade de Votos', ascending=False));

print(" \n ====================== RESULTADO POR ZONA ELEITORAL - (GOVERNADOR) ======================= \n")
print("                              TOTAL de Votos: ", df.query("DS_CARGO == 'GOVERNADOR'")["QT_VOTOS"].sum())

data = [];
for nome in df2.query("DS_CARGO == 'GOVERNADOR'")["NM_VOTAVEL"]:
  print(" \n ====================== RESULTADO DO CANDIDATO " + nome + "======================= \n")
  print("                     Total de Votos do candidato: ", df.query("NM_VOTAVEL == '" + nome + "' ")["QT_VOTOS"].sum())
  for i in numeroZona:
    nomeZona = zonasEleitorais.query("numero_zona == " + str(i))["bairro"];
    nomeZonaString = pd.Series(data=nomeZona)
    data.append([nome, nomeZonaString.values[0], df.query("NR_ZONA == " + str(i) + " & NM_VOTAVEL == '" + nome + "'")["QT_VOTOS"].sum()])
    resultado = pd.DataFrame(data, columns=['Nome', 'Nome Zona Eleitoral', 'Quantidade de Votos'])  
  result = list(resultado['Nome'] == nome)
  print(resultado.loc[result])


resultado.to_csv('governadorZona.csv')

print(" \n ====================== RESULTADO POR ZONA ELEITORAL - (SENADOR) ======================= \n")
print("                              TOTAL de Votos: ", df.query("DS_CARGO == 'GOVERNADOR'")["QT_VOTOS"].sum())

data = [];
for nome in df2.query("DS_CARGO == 'SENADOR'")["NM_VOTAVEL"]:
  print(" \n ====================== RESULTADO DO CANDIDATO " + nome + "======================= \n")
  print("                     Total de Votos do candidato: ", df.query("NM_VOTAVEL == '" + nome + "' ")["QT_VOTOS"].sum())
  for i in numeroZona:
    nomeZona = zonasEleitorais.query("numero_zona == " + str(i))["bairro"];
    nomeZonaString = pd.Series(data=nomeZona)
    data.append([nome, nomeZonaString.values[0], df.query("NR_ZONA == " + str(i) + " & NM_VOTAVEL == '" + nome + "'")["QT_VOTOS"].sum()])
    resultado = pd.DataFrame(data, columns=['Nome', 'Nome Zona Eleitoral', 'Quantidade de Votos'])  
  result = list(resultado['Nome'] == nome)
  print(resultado.loc[result])


resultado.to_csv('senadorZona.csv')

```
